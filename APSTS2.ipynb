>### So just now we have seen that dataset has so many NaN values to fill them with some values we need to do analysis 
How to choose the method to fill the nan value?
>>Here we are using median because if you see in the boxplot the values of K_A_Cases_Reported variable has many outliers.

```python

boxplot = data.boxplot(by="Year", column=['K_A_Cases_Reported'], grid=False)
```

```python
data.fillna(data.median(),inplace=True)
```

```python
data.isna().sum()
```

>We have successfully filled up the NaN value in the data set.

```python
year=data['Year'].unique()
```

```python
label=year
```

```python
df=data.groupby(['Year']).sum()
```

```python
female=df['K_A_Female_Total']
male=df['K_A_Male_Total']
female_cases=female.tolist()
male_cases=male.tolist()
```

```python
def plot_bar_x():
    index = np.arange(len(label))
    Gender=['Female','Male']
    plt.figure(figsize=(12,6))
    plt.bar(index, female_cases)
    plt.bar(index, male_cases)
    plt.xlabel('Years', fontsize=20)
    plt.ylabel('No of people', fontsize=20)
    plt.xticks(index, label, fontsize=20, rotation=0)
    plt.yticks( fontsize=20, rotation=0)
    plt.legend(Gender,loc=2,fontsize=20)
    plt.title('Incidence of Abduction and Kidnapping in each year(2001-2010)', fontsize=20)
    plt.show()
```

>Here if you can see that I have one is a categorical variable(Year) and another one is continuous so to show the relationship between them, I am using bar graph as we know that to show any kind of relationship of two variables we can use a bar graph or line plot if the variables are continuous or categorical.

```python
plot_bar_x()
```

```python
df=data.groupby(['Area_Name']).sum()
```

```python
female=df['K_A_Female_Total']
male=df['K_A_Male_Total']
female_cases=female.tolist()
male_cases=male.tolist()
```

```python
def plot_female_h():
    index = np.arange(len(label))
    Gender=['Female','Male']
    plt.figure(figsize=(22,12))
    plt.barh(index, female_cases)
    plt.barh(index, male_cases)
    plt.ylabel('Years', fontsize=20)
    plt.xlabel('No of people', fontsize=20)
    plt.yticks(index, label, fontsize=20, rotation=0)
    plt.xticks( fontsize=20, rotation=0)
    plt.legend(Gender,loc=4,fontsize=20)
    plt.title('Incidence of Abduction and Kidnapping in each and every state(2001-2010)', fontsize=30)
    plt.show()
    
def plot_male_h():
    index = np.arange(len(label))
    Gender=['Male','Female']
    plt.figure(figsize=(22,12))
    plt.barh(index, male_cases)
    plt.barh(index, female_cases)
    plt.ylabel('Years', fontsize=20)
    plt.xlabel('No of people', fontsize=20)
    plt.yticks(index, label, fontsize=20, rotation=0)
    plt.xticks( fontsize=20, rotation=0)
    plt.legend(Gender,loc=4,fontsize=20)
    plt.title('Incidence of Abduction and Kidnapping in each and every state(2001-2010)', fontsize=30)
    plt.show()    
```

>In the following bar graphs I am showing that which states are having more crime against man and woman.

```python
plot_female_h()
```

If you can see from the above bar graph that almost in all the states crime agaist woman is more comparison to man.

```python
plot_male_h()
```
